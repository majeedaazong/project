```python
x = (dataset['fixed acidity'])
y = (dataset['pH'])
plt.xlabel('Fixed Acidity')
plt.ylabel('pH')
plt.scatter(x,y)
plt.title('Scatter diagram of Fixed acidity against pH')
plt.show()     
   
```

```python
x = (dataset['citric acid'])
y = (dataset['volatile acidity'])
plt.xlabel('citric acid')
plt.ylabel('volatile acidity')
plt.scatter(x,y)
plt.title('Scatter diagram of citric acid against volatile acidity')
plt.show()     
   
```

```python
x = (dataset['free sulfur dioxide'])
y = (dataset['total sulfur dioxide'])
plt.xlabel('free sulfur dioxide')
plt.ylabel('total sulfur dioxide')
plt.scatter(x,y)
plt.title('Scatter diagram of free sulfur dioxide against total sulfur dioxide')
plt.show()     
   
```

From the first graph , we can see that as the fixed acidity increases, he pH decreases.

from the second scatter diagram we can infer that  generally as the amount of citric acid increases, the volatile acidity decreases although there are a few discrepancies or irregularities.

From the third graph generally the amount of free sulfur dioxide increases as total sulfur dioxide also increases.
